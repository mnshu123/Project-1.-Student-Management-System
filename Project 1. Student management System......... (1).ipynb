 # Project- Student Management System:-

```python
student = ['Ravi','Neha','Amit','Rahul','Mnshu']

def showlist():
    if len(student)==0:
       print("list is empty")
    else:
        print("Student list:")
        for i in range(len(student)):
            print(student[i])


def addlist():
    a = input("enter student name to Add:")
    student.append(a)
    print("Student added Successfully:")


def removelist():
    a = input("enter Student name to remove:")
    for i in range(len(student)):
        if student[i]==a:
            student.remove(a)
            print("Student removed Successfully:")
            break
    else:
        print("student not found")



def searchlist():
    a = input("enter student name to search:")
    for i in range(len(student)):
        if student[i] == a:
            print("Student is Present")
            break
    else:
        print("Student not present")



while True:
    print("/n--- Student Management System ---")
    print("1. Show Student List")
    print("2. Add Student")
    print("3. Remove Student")
    print("4. Search Student")
    print("5. Exit")

    ch = int(input("Enter your choice:"))
    if ch ==1:
        showlist()
    elif ch ==2:
        addlist()
    elif ch==3:
        removelist()
    elif ch==4:
        searchlist()
    elif ch==5:
        print("Thank You")
        break
    else:
        print("Invalid Choice")
        
```
